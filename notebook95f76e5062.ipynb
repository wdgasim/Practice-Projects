```python
df['travel_time'].unique()
```

```python
def categorize_travel_time(time):
    str_time = str(time)
    
    if str_time == '<15 min':
        return'Very Short'
    elif str_time == '15-30 min':
        return'Short'
    elif str_time == '30-60 min':
        return"Medium"
    else:
        return"Long"
```

```python
df['travel_time'] = df['travel_time'].apply(lambda x: categorize_travel_time(x) )
```

I have changed the values of 'travel_time' column to:
- Very Short for <15 min.
- Short for 15-30 min.
- Medium for 30-60 min.
- Long for >60 min.

## Multivariate

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['study_hours'], df['attendance_percentage'], df['overall_score'],
                     c=df['overall_score'],  
                     cmap='coolwarm',
                     s=50,
                     alpha=0.7,
                     depthshade=True)

ax.set_xlabel('Study Hours', labelpad=15)
ax.set_ylabel('Attendance Percentage', labelpad=15)
ax.set_zlabel('Overall Score', labelpad=15)
plt.colorbar(scatter, label='Overall Score', pad=0.1)
plt.title('3D Scatter Plot')
plt.show()
```

The 3D scatter plot demonstrates a positive linear relationship between study hours, attendance percentage, and overall score. As both study hours and attendance percentage increase, overall scores show a corresponding upward trend, suggesting that these two factors collectively contribute to better academic performance.

```python
corr_matrix = df[['study_hours', 'attendance_percentage', 'math_score', 'science_score', 'english_score', 'overall_score']].corr() 

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Interpretation & Implications:

- Study hours matter more than attendance for predicting academic success in this dataset.
- Subject scores are interrelated, suggesting students who do well in one subject tend to do well in others.
- Attendance percentage has minimal direct linear impact on scores  perhaps other factors (like study quality, prior knowledge, or motivation) mediate its effect.
