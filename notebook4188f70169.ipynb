```python
sns.countplot(x = "Transmission", data = df)
plt.title("Comparison Between Transmission Types")
plt.xlabel("Transmission Types")
plt.grid(True, alpha= 1)
plt.show()
```

```python
sns.lineplot(x = 'Year', y = 'Price', data = df)
```

```python
plt.figure(figsize = (8,6))
car = df["Brand"].value_counts()
plt.pie(car, labels = car.index , autopct = '%1.1f%%', startangle = 90)
plt.title("Brand-wise Car Count in Dataset")
plt.axis("equal")
plt.tight_layout()
plt.grid(True , alpha = 0.9999)
plt.show()
```

```python
plt.figure(figsize = (8,6))
v_1 = df["Model"].value_counts()
v_1.plot(kind = 'bar', color = 'steelblue')
plt.title("Most Common Car Models in Dataset")
plt.xlabel("Car Model")
plt.ylabel("Number of Cars")
plt.tight_layout() 
plt.show()
```
